### Part 1

```python
import numpy as np
from loguru import logger

with open('../data/day10.txt') as f:
    input_data = f.read().split('\n')

    
results = []
X = 1
for command in input_data:
    if command == 'noop':
        results.append(X)
    elif command.startswith('addx'):
        results.append(X)
        results.append(X)
        X += int(command.split()[-1])
    else: 
        logger.error('Command not implemented.')

        
        
sum([i * results[i-1] for i in [20, 60, 100, 140, 180, 220]])
```

### Part 2

```python
result = []
current_crt = []
for cycle in results:
    if len(current_crt) == 40:
        print(''.join(current_crt))
        result.append(''.join(current_crt))
        current_crt.clear()
    if len(current_crt) in [cycle -1, cycle, cycle + 1]:
        current_crt.append('0')
    else:
        current_crt.append('.')
print(''.join(current_crt))    
```
